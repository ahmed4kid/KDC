## This notebook plots the KN lightcurves

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

min_x = 0
max_x = 10
min_y = 20
max_y = 30
```

```python
# These functions plot band graphs
```

```python
def plot_u_band(u, days_u, file_name):
    for x in range(len(data2)):
        u = data2['u'][x]
        days_u = data2['time'][x]
        scatter = plt.plot(days_u, u, '.g-') 
    plt.title('u-band ' + file_name)
    plt.xlabel('Days After')
    plt.ylabel('u-band magnitude')
    plt.xlim(min_x,max_x)
    plt.ylim(20,30)
    plt.gca().invert_yaxis()
    plt.show()
    return u, days_u
```

```python
def plot_g_band(g, days_g, file_name):
    for x in range(len(data2)):
        g = data2['g'][x]
        days_g = data2['time'][x]
        plt.figure(0)
        scatter = plt.plot(days_g, g, '.b-')
    plt.figure(0)
    plt.title('g-band ' + file_name)
    plt.xlabel('Days After')
    plt.ylabel('g-band magnitude')
    plt.xlim(min_x, max_x)
    plt.ylim(20,30)
    plt.gca().invert_yaxis()
    plt.show()
    return g, days_g
```

```python
def plot_r_band(r, days_r, file_name):
    for x in range(len(data2)):
        r = data2['r'][x]
        days_r = data2['time'][x]
        scatter = plt.plot(days_r, r, '.g-') 
    plt.title('r-band ' + file_name)
    plt.xlabel('Days After')
    plt.ylabel('r-band magnitude')
    plt.xlim(min_x, max_x)
    plt.ylim(min_y,30)
    plt.gca().invert_yaxis()
    plt.show()
    return r, days_r
```

```python
def plot_i_band(i, days_i, file_name):
    for x in range(len(data2)):
        i = data2['i'][x]
        days_i = data2['time'][x]
        scatter = plt.plot(days_i, i, '.r-') 
    plt.title('i-band ' + file_name)
    plt.xlabel('Days After')
    plt.ylabel('i-band magnitude')
    plt.xlim(min_x,max_x)
    plt.ylim(min_y, max_y)
    plt.gca().invert_yaxis()
    plt.show()
    return i, days_i
```

```python
def plot_z_band(z, days_z, file_name):
    for x in range(len(data2)):
        z = data2['z'][x]
        days_z = data2['time'][x]
        scatter = plt.plot(days_z, z, '.y-') 
    plt.title('z-band ' + file_name)
    plt.xlabel('Days After')
    plt.ylabel('z-band magnitude')
    plt.xlim(min_x, max_x)
    plt.ylim(min_y, max_y)
    plt.gca().invert_yaxis()
    plt.show()
    return z, days_z
```

```python
def plot_y_band(y, days_y, file_name):
    for x in range(len(data2)):
        y = data2['y'][x]
        days_y = data2['time'][x]
        scatter = plt.plot(days_y, y, '.b-') 
    plt.title('y-band ' + file_name)
    plt.xlabel('Days After')
    plt.ylabel('y-band magnitude')
    plt.xlim(min_x, max_x)
    plt.ylim(min_y, max_y)
    plt.gca().invert_yaxis()
    plt.show()
    return y, days_y
```

```python
def plot_joint_bands(u, g, r, i, z, y, days_u, days_g, days_r, days_i, days_z, days_y, file_name):
    # plot with all bands 
    # currently not plotting - variables print out with correct values  
    scatter = plt.plot(days_u, u, '.b', label = "u") 
    scatter = plt.plot(days_g, g, '.g-', label = "g")
    scatter = plt.plot(days_r, r, '.r-', label = "r")
    scatter = plt.plot(days_i, i, '.m-', label = "i")
    scatter = plt.plot(days_z, z, '.k-', label = "z")
    scatter = plt.plot(days_y, y, '.y-', label = "y") 
    # the graph sets up correctly 
    plt.title('Joint Band Plot for ' + file_name)
    plt.xlabel('Days After')
    plt.ylabel('Magnitude')
    plt.xlim(min_x, max_x)
    plt.ylim(min_y, max_y)
    plt.gca().invert_yaxis()
    plt.legend()
    plt.show()
```

```python
# Read in files, plot all individual band graphs, plot joint graph
```
