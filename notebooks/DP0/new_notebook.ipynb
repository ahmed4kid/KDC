## This notebook plots Maximum Magnitude vs. System Number for the Light Curve Files


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Read in light_curve_ALL_* files and find the brightest magnitudes

```python
file_path = "/home/annakhalid1/KDC/notebooks/output/light_curve_ALL_*.txt"

#/home/annakhalid1/KDC
#KDC/notebooks/output/light_curve_ALL_0.txt

brightest_magnitudes = []
system_numbers = []
#keep track of the points for the plot


    """
This cell iterates through all 328 light curve files by using a for loop
and locate & stores the value of the maximum magnitude for r-band in each file 

    """

for i in range(329):
    file_name = file_path.replace("*", str(i))
    #data = np.loadtxt(file_name) 

    data = np.loadtxt(file_name, delimiter=',', skiprows=1)
    #skips the header row u,g,r,i,z,y
    
    brightest_mag = np.min(data[:, 2])  
    #finding the max magnitude for r-band (at 2nd index) (3rd index is actually i)
    
    system_numbers.append(i)
    #appends the value of variable i to the system_numbers list (it stores the system number for each iteration of the loop.)

    brightest_magnitudes.append(brightest_mag) 
    #does the same for brightest_mag calulation above



#printing out to make sure the loop works
#    print("light_curve_ALL_",i)
#    print("Brightest Magnitude (r-band):", brightest_mag)
#    print("")
```

## Plotting r-band index vs brightest magnitude

```python
plt.plot(system_numbers, brightest_magnitudes, '.r')# color="red", marker="o")
plt.xlabel("System Number")
plt.ylabel("Brightest Magnitude (r-band)")
plt.title("Brightest Magnitude vs. System Number")
plt.gca().invert_yaxis()
plt.show()


#plt.xlim(0,9)
#plt.ylim(16,28)
#plt.gca().invert_yaxis()
#plt.legend()
#plt.show()
```
