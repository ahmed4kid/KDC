## This notebook plots the KN lightcurves

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data2 = pd.read_csv('output/light_curve.txt')
#timers = pd.read_csv('output/time.txt')

#data2 = pd.concat([mags['u'],mags['g'],mags['r'],mags['i'],mags['z'],mags['y'],timers['day_after']], axis=1)

#data2 = pd.read_csv('input/magnitudes.txt')
#days = data2['day_after']

#days_u = data.loc[data['u'] > -9999]['day_after']
#u = data.loc[data['u'] > -9999]['u']

u = data2['u']
g = data2['g']
r = data2['r']
iii = data2['i']
z = data2['z']
Y = data2['y']
days = data2['time']
```

## Read in the real data, sort by filter

```python
data = pd.read_csv('input/gw170817.data', delim_whitespace=True)

#mask = data>-9999
#data = data[mask]

#u = data[data['u']>-9999]
#g = data[data['g']>-9999]
#r = data[data['r']>-9999]
#i = data[data['i']>-9999]
#z = data[data['z']>-9999]
#Y = data[data['y']>-9999]



i_band_mag = data.loc[data['Band'] == 'i']['Mag']
i_band_time = data.loc[data['Band'] == 'i']['MJD']
z_band_mag = data.loc[data['Band'] == 'z']['Mag']
z_band_time = data.loc[data['Band'] == 'z']['MJD']
Y_band_mag = data.loc[data['Band'] == 'Y']['Mag']
Y_band_time = data.loc[data['Band'] == 'Y']['MJD']
r_band_mag = data.loc[data['Band'] == 'r']['Mag']
r_band_time = data.loc[data['Band'] == 'r']['MJD']
g_band_mag = data.loc[data['Band'] == 'g']['Mag']
g_band_time = data.loc[data['Band'] == 'g']['MJD']
u_band_mag = data.loc[data['Band'] == 'u']['Mag']
u_band_time = data.loc[data['Band'] == 'u']['MJD']
```

## Make the difference plot

```python
u_time = u_band_time-zero
print(len(u_time))
g_time = g_band_time-zero
print(len(g_time))
r_time = r_band_time-zero
print(len(r_time))
i_time = i_band_time-zero
print(len(i_time))
z_time = z_band_time-zero
print(len(z_time))
Y_time = Y_band_time-zero
print(len(Y_time))

sim_u = [0] * len(u_time)
real_u = [0] * len(u_time)
sim_g = [0] * len(g_time)
real_g = [0] * len(g_time)
sim_r = [0] * len(r_time)
real_r = [0] * len(r_time)
sim_i = [0] * len(i_time)
real_i = [0] * len(i_time)
sim_z = [0] * len(z_time)
real_z = [0] * len(z_time)
sim_Y = [0] * len(Y_time)
real_Y = [0] * len(Y_time)

uu = [0] * len(u_time)
gg =[0] * len(g_time)
rr =[0] * len(r_time)
ii =[0] * len(i_time)
zz =[0] * len(z_time)
YY =[0] * len(Y_time)

u_real =[0] * len(u_time)
g_real =[0] * len(g_time)
r_real =[0] * len(r_time)
i_real =[0] * len(i_time)
z_real =[0] * len(z_time)
Y_real =[0] * len(Y_time)

delta_u =[0] * len(u_time)
delta_g =[0] * len(g_time)
delta_r =[0] * len(r_time)
delta_ii =[0] * len(i_time)
delta_z =[0] * len(z_time)
delta_Y =[0] * len(Y_time)

#print(u_band_mag) #4
#print(g_band_mag) #6
#print(r_band_mag) #8
#print(i_band_mag) #9
#print(z_band_mag) #11
#print(Y_band_mag) #10

for x in range (0, len(u_time)):
    diff = abs(u_time.iloc[x] - days)
    this = np.where(diff == min(diff))
    real_u[x] = u_time.iloc[x]
    sim_u[x] = float(days.iloc[this].to_string(index=False))
    uu[x] = u_band_mag.iloc[x]
    u_real[x] = float(u.iloc[this].to_string(index=False))
    delta_u[x] = u_real[x] - uu[x]

for x in range (0, len(g_time)):
    diff = abs(g_time.iloc[x] - days)
    this = np.where(diff == min(diff))
    real_g[x] = g_time.iloc[x]
    sim_g[x] = float(days.iloc[this].to_string(index=False))
    gg[x] = g_band_mag.iloc[x]
    g_real[x] = float(g.iloc[this].to_string(index=False))
    delta_g[x] = g_real[x] - gg[x]
  
for x in range (0, len(r_time)):
    diff = abs(r_time.iloc[x] - days)
    this = np.where(diff == min(diff))
    real_r[x] = r_time.iloc[x]
    sim_r[x] = float(days.iloc[this].to_string(index=False))
    rr[x] = r_band_mag.iloc[x]
    r_real[x] = float(r.iloc[this].to_string(index=False))
    delta_r[x] = r_real[x] - rr[x]

for x in range (0, len(i_time)):
    diff = abs(i_time.iloc[x] - days)
    this = np.where(diff == min(diff))
    real_i[x] = i_time.iloc[x]
    sim_i[x] = float(days.iloc[this].to_string(index=False))
    ii[x] = i_band_mag.iloc[x]
    i_real[x] = float(iii.iloc[this].to_string(index=False)) 
    delta_ii[x] = i_real[x] - ii[x]
    

for x in range (0, len(z_time)):
    diff = abs(z_time.iloc[x] - days)
    this = np.where(diff == min(diff))
    real_z[x] = z_time.iloc[x]
    sim_z[x] = float(days.iloc[this].to_string(index=False))
    zz[x] = z_band_mag.iloc[x]
    z_real[x] = float(z.iloc[this].to_string(index=False))
    delta_z[x] = z_real[x] - zz[x]

for x in range (0, len(Y_time)):
    diff = abs(Y_time.iloc[x] - days)
    this = np.where(diff == min(diff))
    real_Y[x] = Y_time.iloc[x]
    sim_Y[x] = float(days.iloc[this].to_string(index=False))
    YY[x] = Y_band_mag.iloc[x]
    Y_real[x] = float(Y.iloc[this].to_string(index=False))
    delta_Y[x] = Y_real[x] - YY[x]

print(uu)
print(u_real)
print(gg)
print(g_real)
print(rr)
print(r_real)
print(ii)
print(i_real)
print(zz)
print(z_real)
print(YY)
print(Y_real)

zero_time = [-5,100]
zeroo = [0,0]

f = plt.figure()
f.set_figwidth(10)
f.set_figheight(10)

scatter = plt.plot(real_u, delta_u, '.b-', label = "u") 
scatter = plt.plot(real_g, delta_g, '.g-', label = "g")
scatter = plt.plot(real_r, delta_r, '.r-', label = "r")
scatter = plt.plot(real_i, delta_ii,'.m-', label = "i")
scatter = plt.plot(real_z, delta_z,'.k-', label = "z")
scatter = plt.plot(real_Y, delta_Y,'.y-', label = "y") 
scatter = plt.plot(zero_time, zeroo, '.k-')
plt.title('Difference between real and simulated')
plt.xlabel('Days after GW')
plt.ylabel('Real - Simulated Magnitude')
plt.xlim(0,9)
#plt.ylim(14,30)
plt.gca().invert_yaxis()
plt.legend()
plt.show()


```

## This is a first attempt at a Gaussian distribution to predict redshifts

```python
mu, sigma = 0.1, 0.03 # mean and standard deviation
s = np.random.normal(mu, sigma, 10)
print(s)
abs(mu - np.mean(s))
0.0  # may vary
abs(sigma - np.std(s, ddof=1))
0.1  # may vary
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```
