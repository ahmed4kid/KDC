## This notebook plots the KN lightcurves

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Read the data, sort it by filter

```python
data = pd.read_csv('input/magnitudes.txt')

#mask = data>-9999
#data = data[mask]

#u = data[data['u']>-9999]
#g = data[data['g']>-9999]
#r = data[data['r']>-9999]
#i = data[data['i']>-9999]
#z = data[data['z']>-9999]
#Y = data[data['y']>-9999]

days = data['day_after']

#days_u = data.loc[data['u'] > -9999]['day_after']
#u = data.loc[data['u'] > -9999]['u']

u = data['u']
g = data['g']
r = data['r']
i = data['i']
z = data['z']
Y = data['y']
```

## Make the plots

```python
scatter = plt.plot(days, u, '.g') 
plt.title('u-band')
plt.xlabel('Days After')
plt.ylabel('u-band magnitude')
plt.xlim(0,25)
plt.ylim(15,25)
plt.gca().invert_yaxis()
plt.show()

scatter = plt.plot(days, g, '.b-') 
plt.title('g-band')
plt.xlabel('Days After')
plt.ylabel('g-band magnitude')
plt.xlim(0,25)
plt.ylim(15,25)
plt.gca().invert_yaxis()
plt.show()

scatter = plt.plot(days, r, '.g-') 
plt.title('r-band')
plt.xlabel('Days After')
plt.ylabel('r-band magnitude')
plt.xlim(0,25)
plt.ylim(15,25)
plt.gca().invert_yaxis()
plt.show()

scatter = plt.plot(days, i, '.r-') 
plt.title('i-band')
plt.xlabel('Days After')
plt.ylabel('i-band magnitude')
plt.xlim(0,25)
plt.ylim(15,25)
plt.gca().invert_yaxis()
plt.show()

scatter = plt.plot(days, z, '.y-') 
plt.title('z-band')
plt.xlabel('Days After')
plt.ylabel('z-band magnitude')
plt.xlim(0,25)
plt.ylim(15,25)
plt.gca().invert_yaxis()
plt.show()

scatter = plt.plot(days, Y, '.b-') 
plt.title('y-band')
plt.xlabel('Days After')
plt.ylabel('y-band magnitude')
plt.xlim(0,25)
plt.ylim(15,25)
plt.gca().invert_yaxis()
plt.show()
```

```python
scatter = plt.plot(days, u, '.g', label = "u") 
scatter = plt.plot(days, g, label = "g")
scatter = plt.plot(days, r, label = "r")
scatter = plt.plot(days, i, label = "i")
scatter = plt.plot(days, z, label = "z")
scatter = plt.plot(days, Y, label = "y") 
plt.title('All Bands')
plt.xlabel('Days After')
plt.ylabel('Band Magnitude')
plt.xlim(0,25)
plt.gca().invert_yaxis()
plt.legend()
plt.show()
```
