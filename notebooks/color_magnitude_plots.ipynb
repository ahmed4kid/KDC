```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
file_path = 'input/magnitudes.csv'
data = pd.read_csv(file_path)
```

```python
# set up plot boundareies
min_x = 10
max_x = 25
min_y = -10
max_y = 10
```

```python
def plot_r_vs_ri():
    for x in range(len(data)):
        color = data['rApMag'][x]
        magnitude = data['r_i'][x]
        scatter = plt.plot(color, magnitude, '.g-')
    # plot ngc 4993
    scatter = plt.plot(12.80, 0.45, '.b-')
    plt.title('rApMag vs. r - i')
    plt.xlabel('Color')
    plt.ylabel('Magnitude')
    plt.xlim(min_x, max_x)
    plt.ylim(min_y, max_y)
    plt.show()
```

```python
def plot_g_vs_gr():
    for x in range(len(data)):
        color = data['gApMag'][x]
        magnitude = data['g_r'][x]
        scatter = plt.plot(color, magnitude, '.g-')
    # plot ngc 4993
    scatter = plt.plot(11.90, 0.90, '.b-')
    plt.title('gApMag vs. g - r')
    plt.xlabel('Color')
    plt.ylabel('Magnitude')
    plt.xlim(min_x, max_x)
    plt.ylim(min_y, max_y)
    plt.show()
```

```python
plot_r_vs_ri()
plot_g_vs_gr()
```
